```python
# adding year and phenology info
anthesis_2017_edit["year"] = 2017
anthesis_2017_edit["phenology"] = "anthesis"
veraison_2017["year"] = 2017
veraison_2017["phenology"] = "veraison"
harvest_2017["year"] = 2017
harvest_2017["phenology"] = "harvest"

anthesis_2018["year"] = 2018
anthesis_2018["phenology"] = "anthesis"
veraison_2018["year"] = 2018
veraison_2018["phenology"] = "veraison"
harvest_2018_edit["year"] = 2018
harvest_2018_edit["phenology"] = "harvest"

anthesis_2019["year"] = 2019
anthesis_2019["phenology"] = "anthesis"
veraison_2019_edit["year"] = 2019
veraison_2019_edit["phenology"] = "veraison"
harvest_2019["year"] = 2019
harvest_2019["phenology"] = "harvest"
```

```python
# merge all datasets
vitis_ionomics_full = pd.concat([anthesis_2017_edit, veraison_2017, harvest_2017, anthesis_2018, veraison_2018, harvest_2018_edit, anthesis_2019, veraison_2019_edit, harvest_2019], sort=False)
```

```python
#degrade sample name into parts

# meta store
row = []
cell = []
replicate = []
leafpos = []
coords = []

# anthesis 2017 data has sample name formatted differently than all other timepoints/years (no _ separation)
def ant2017(samp):
    if len(samp) == 4:
        rowid = samp[0]
        row.append(samp[0])
    else:
        rowid = samp[0:2]
        row.append(samp[0:2])
    cellid = samp[-3]
    cell.append(samp[-3])
    replicate.append(samp[-2])
    leafpos.append(samp[-1])
    coords.append(str(cellid)+str(rowid))
    
# annoyingly, anthesis 2019 and veraison 2019 are also different. (no _ separation between middle two)
def antver2019(samp):
    samp = samp.split("_")
    rowid = samp[0]
    row.append(samp[0])
    cellid = samp[1][0:1]
    cell.append(samp[1][0:1])
    replicate.append(samp[1][1:2])
    leafpos.append(samp[2])
    coords.append(str(cellid)+str(rowid))

# all other timepoints/years sample name degradation (_ separation)
def name_degrade(samp):
    samp = samp.split("_")
    rowid = samp[0]
    row.append(samp[0])
    cellid = samp[1]
    cell.append(samp[1])
    replicate.append(samp[2])
    leafpos.append(samp[3])
    coords.append(str(cellid)+str(rowid))
```

```python
for samp in vitis_ionomics_full["sample"]:
    if "_" not in samp:
        ant2017(samp)
    elif samp.count('_') == 2:
        antver2019(samp)
    else:
        name_degrade(samp)
```

```python
vitis_ionomics_full["row"] = row
vitis_ionomics_full["cell"] = cell
vitis_ionomics_full["replicate"] = replicate
vitis_ionomics_full["leafpos"] = leafpos
vitis_ionomics_full["coords"] = coords
```

```python
# some horrible, terrible person accidentally put 1O instead of 10. 
vitis_ionomics_full.iat[5134,35] = '10'
vitis_ionomics_full.iat[5134,39] = 'G10'
```

```python
# Info on irrigation treatment
treatment = []
i = 0
for row in vitis_ionomics_full["row"]:
    row = int(row)
    if (row == 16 or row == 12 or row ==10):
        treatment.append("Full")
    elif (row == 15 or row == 13 or row ==8):
        treatment.append("None")
    else:
        treatment.append("RDI")
    i += 1
vitis_ionomics_full["treatment"] = treatment
```
